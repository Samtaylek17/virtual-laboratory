## Starting the Server

Run:
docker-compose up

Stop:
docker-compose down

## Quiz

The quiz for the server is stored in "/srv/nbgrader/exchange/quiz"

QUIZ PATH: /srv/nbgrader/exchange/quiz::11/30/2018 01:15PM::120

- It contains the date, time and duration of the quiz
- Always restart after making any changes

## Attendance

The attendance is stored in "/srv/nbgrader/exchange/quiz"


## Result

Quiz result is in "/srv/nbgrader/exchange/quiz.result"
