## Linearity Property

A linear circuit is one whose output is linearly related (or directly proportional) to its input.

Assume that $V_{o}$ = 1V and use linearity to calculate the actual value
of $V_{o}$ in the circuit of Fig. 4.5

```python
from IPython.display import display, Markdown, clear_output
import ipywidgets as widgets
import schemdraw
import schemdraw.elements as elm

d = schemdraw.Drawing()
d += elm.Resistor().right().label('12Ω')
d += elm.Resistor().down().label('8Ω').label(('+', '$V_o$', '-'), loc='bottom')
d += elm.Line().left().label('Fig. 4.5')
d += elm.Resistor().up().label('5Ω')
d += elm.Line().left()
d += elm.SourceV().down().reverse().label('30V')
d += elm.Line().right()
d
```
